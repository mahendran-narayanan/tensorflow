From https://www.tensorflow.org/guide/keras/overview


This notebook contains basic implementation of neural network with keras using random input data.

```python
import tensorflow as tf
import numpy as np
```

```python
data = np.random.random((1000,32))
label = np.random.random((1000,10))
label
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64,activation='relu',input_shape=(32,)),
    tf.keras.layers.Dense(64,activation='relu'),
    tf.keras.layers.Dense(64,activation='relu'),
    tf.keras.layers.Dense(10)
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.01),
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])
```

```python
model.fit(data,label,epochs=203,batch_size=32)
```

```python
vdata = np.random.random((1000,32))
vlabel = np.random.random((1000,10))
model.fit(data,label,epochs=10,batch_size=32,
         validation_data=(vdata,vlabel))
```

```python
rdata = np.random.random((1000,32))
rlabel = np.random.random((1000,10))
model.evaluate(rdata,rlabel,batch_size=32)
```
